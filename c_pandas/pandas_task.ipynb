### 통계값 확인 및 데이터 처리
- describe()로 수치 데이터의 통계값들을 확인해 봅니다.
- 결측 데이터가 있는지 확인하여 처리합니다.
- 각 수치 데이터마다 하위 25%(Q1), 상위 25%(Q3) 값 및 iqr 값과 이를 기준으로 한 정상 범위를 확인해 봅니다.
- 정상 범위에서 벗어난 이상치가 있다면 삭제합니다.

```python
# describe()로 데이터 수, 평균, 표준편차, 최솟값, 최댓값, 사분위 분포도 확인
display(happiness_df.describe().T)
```

```python
# mode()로 feature 별 최빈값 확인
happiness_df.mode()
```

```python
# 결측 데이터가 있는지 확인
print(happiness_df.isna().sum())
```

```python
# healthy_life_expectancy feature의 결측치 1개를 해당 feature의 평균값으로 대체
healthy_mean = happiness_df['healthy_life_expectancy'].mean()
happiness_df['healthy_life_expectancy'] = happiness_df['healthy_life_expectancy'].fillna(healthy_mean)
print(happiness_df.isna().sum())
```

```python
# Q1, Q3, iqr 값을 구하고 이를 기준으로 정상범위를 구해 이상치 제거하기
# happiness_score
score_Q1 = happiness_df.describe().T.loc['happiness_score', '25%']
score_Q3 = happiness_df.describe().T.loc['happiness_score', '75%']
score_iqr_value = score_Q3 - score_Q1
score_lower_bound = score_Q1 - 1.5 * score_iqr_value
score_upper_bound = score_Q3 + 1.5 * score_iqr_value
print(f'happiness_score\n하위 25%: {score_Q1}, 상위 25%: {score_Q3}')
print(f'iqr값: {score_iqr_value}')
print(f'정상 범위: {score_lower_bound} ~ {score_upper_bound}')

print(f'happiness_score 이상치 제거 전 행의 개수: {len(happiness_df)}')
happiness_df = happiness_df[(score_lower_bound <= happiness_df['happiness_score']) & (happiness_df['happiness_score'] <= score_upper_bound)]
print(f'happiness_score 이상치 제거 후 행의 개수: {len(happiness_df)}')
```

```python
# gdp_per_capita 
gdp_Q1 = happiness_df.describe().T.loc['gdp_per_capita', '25%']
gdp_Q3 = happiness_df.describe().T.loc['gdp_per_capita', '75%']
gdp_iqr_value = gdp_Q3 - gdp_Q1
gdp_lower_bound = gdp_Q1 - 1.5 * gdp_iqr_value
gdp_upper_bound = gdp_Q3 + 1.5 * gdp_iqr_value
print(f'gdp_per_capita\n하위 25%: {gdp_Q1}, 상위 25%: {gdp_Q3}')
print(f'iqr값: {gdp_iqr_value}')
print(f'정상 범위: {gdp_lower_bound} ~ {gdp_upper_bound}')

print(f'gdp_per_capita 이상치 제거 전 행의 개수: {len(happiness_df)}')
happiness_df = happiness_df[(gdp_lower_bound <= happiness_df['gdp_per_capita']) & (happiness_df['gdp_per_capita'] <= gdp_upper_bound)]
print(f'gdp_per_capita 이상치 제거 후 행의 개수: {len(happiness_df)}')
```

```python
# social_support
social_Q1 = happiness_df.describe().T.loc['social_support', '25%']
social_Q3 = happiness_df.describe().T.loc['social_support', '75%']
social_iqr_value = social_Q3 - social_Q1
social_lower_bound = social_Q1 - 1.5 * social_iqr_value
social_upper_bound = social_Q3 + 1.5 * social_iqr_value
print(f'social_support\n하위 25%: {social_Q1}, 상위 25%: {social_Q3}')
print(f'iqr값: {social_iqr_value}')
print(f'정상 범위: {social_lower_bound} ~ {social_upper_bound}')

print(f'social_support 이상치 제거 전 행의 개수: {len(happiness_df)}')
happiness_df = happiness_df[(social_lower_bound <= happiness_df['social_support']) & (happiness_df['social_support'] <= social_upper_bound)]
print(f'social_support 이상치 제거 후 행의 개수: {len(happiness_df)}')
```

```python
# 각 요인과 happiness_score(행복점수)의 상관관계를 DataFrame.corr()을 통해 구합니다.
correlation_df = pd.DataFrame(happiness_df[['happiness_score', 'gdp_per_capita', 'social_support', 
              'healthy_life_expectancy', 'freedom_to_make_life_choices', 
              'generosity', 'perceptions_of_corruption']].corr()['happiness_score']).rename(columns={'happiness_score': 'correlation_strength'})
display(correlation_df)
```

```python
# 요인별 상관관계의 강도를 시각화합니다.
import matplotlib.pyplot as plt
correlation_plot = correlation_df.plot(title='Correlation', y='correlation_strength', fontsize='10', kind='bar', grid=True)
correlation_plot.set_xticklabels(correlation_df.index, rotation=45)
for i, value in enumerate(correlation_df['correlation_strength']):
    plt.text(i, value + 0.01, str(round(value, 2)), ha='center', fontsize='10')
plt.show()
```

#### 2023년도 경제·사회·정치적 요인과 행복지수의 상관관계에 대한 보고서
- 총 6가지의 경제·사회·정치적 요인에 대하여 행복점수(happiness_score)와의 관계를 시각화하고 상관관계를 구한 결과,
  각 요인을 상관관계의 강도(strength)에 대한 일반적인 기준에 따라 3가지 그룹으로 분류할 수 있었다.
  
- 상관관계의 강도가 0.8 이상인 "강한 상관관계(Strong correlation)" 그룹에는 1인당 국내총생산(GDP)과 사회적 지지(social support)가 속한다.  
  해당 그룹 내의 요인들과 행복 점수의 상관관계 강도는 각각 약 0.83, 0.81로 강한 양의 상관관계를 띤다.  
  이는 이 요인들의 수치와 행복 점수가 비례함을 시사한다.

- 상관관계의 강도가 0.8에서 0.4 사이인 "중간 상관관계(Moderate correlation)" 그룹에는 건강한 기대수명(healthy life expectancy)과  
  삶의 선택에 대한 자유로움의 척도(freedom to make life choices)가 속한다.  
  이 그룹의 요인들과 행복 점수 사이의 상관관계 강도는 각각 약 0.77, 0.53으로, 중간 정도의 양의 상관관계를 띤다.  
  건강한 기대수명 요인은 그룹 내 다른 요인보다 행복 점수에 보다 강한 영향을 미치고 있음을 확인할 수 있다.

- 상관관계의 강도가 0.4 이하인 "약한 상관관계(Weak correlation)" 그룹에는 관대함(generosity)과    
  국가 내 부패에 대한 인식의 척도(perceptions of corruption), 즉 정치적 요인이 속한다.  
  전자와 행복 점수의 상관관계 강도는 약 -0.06으로 약한 음의 상관관계를 보이며, 후자는 약 0.29로 약한 양의 상관관계를 보인다.  
  따라서 해당 그룹 내의 요인들은 행복 점수에 미치는 영향이 작을 것으로 예상된다.  

📌 요약하자면, 경제·사회적 요인의 수치는 일반적으로 행복 점수와 비례하지만, 정치적 요인은 행복 점수에 미치는 영향이 상대적으로 작다.  
이처럼 정치적 요인이 행복 점수에 미치는 영향이 제한적임을 고려할 때,  
국민들은 더 많은 관심과 참여를 통해 정치적 결정에 더 큰 영향력을 행사할 필요가 있다.
